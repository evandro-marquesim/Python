# _detalhando_

```python
plt.title('Pre√ßo do combustivel com o passar dos anos (em d√≥lares americanos)', fontdict={'fontweight': 'bold', 'fontsize':25})



eua = plt.plot(banco_gas.Year, banco_gas.USA, 'b.-', label = 'EUA' ) # (x,y)
can = plt.plot(banco_gas.Year, banco_gas.Canada, 'r.-', label = 'Canada') # multiplas linhas
jap = plt.plot(banco_gas.Year, banco_gas.Japan, 'g.-', label = 'Jap√£o') #modelo: cor formato. green, point, trace 

plt.xticks(banco_gas.Year) # cria uma marca√ß√£o ou tick por ano 

plt.xlabel('Ano')
plt.ylabel('Pre√ßo da gasolina em d√≥lares americanos')

plt.legend() # monta a legenda com rela√ß√£o aos labels das vari√°veis

plt.savefig('Pre√ßo_gasolina_paises.png', dpi=300)

plt.show() 
```
